```python
# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", lambda x: '%.2f' % x)

# %%
df = pd.read_csv("online_retail.csv", encoding="ISO-8859-1")

# %%
df.info()

# %%
df.head()

# %%
print("Satır sayısı:", df.shape[0])
print("Sütun sayısı:", df.shape[1])

# %%
df.describe()

# %%
df.describe(include="object")

# %%
df_tr = df.rename(columns={
    "InvoiceNo": "FaturaNo",
    "StockCode": "UrunKodu",
    "Description": "UrunAdi",
    "Quantity": "Adet",
    "InvoiceDate": "FaturaTarihi",
    "UnitPrice": "BirimFiyat",
    "CustomerID": "MusteriID",
    "Country": "Ulke"
})


# %%
df.describe()

# %%
df.isnull().sum().sort_values(ascending=False)

# %%
df = df.dropna(subset=["CustomerID"])

# %%
product_map = (
    df.dropna(subset=["Description"])
      .groupby("StockCode")["Description"]
      .agg(lambda x: x.mode().iloc[0])
)


# %%
df["Description"] = df.apply(
    lambda row: product_map[row["StockCode"]]
    if pd.isnull(row["Description"]) and row["StockCode"] in product_map
    else row["Description"],
    axis=1
)


# %%
df.isnull().sum()


# %%
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# %%
df[df["Quantity"]  <= 0].shape[0]

# %%
df = df[df["Quantity"] > 0]


# %%
df[df["UnitPrice"] <= 0].shape[0]
df = df[df["UnitPrice"] > 0]

# %%
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

# %%
df["Year"] = df["InvoiceDate"].dt.year
df["Month"] = df["InvoiceDate"].dt.month
df["Day"] = df["InvoiceDate"].dt.day
df["weekday"] = df["InvoiceDate"].dt.day_name()

# %%
df.info()
df.describe()


# %%
df.to_csv("online_retail_cleaned.csv", index=False)


# %% [markdown]
# notbook 2 eda trend analizi

# %%
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("online_retail_cleaned.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# %%
total_revenue = df["TotalPrice"].sum()
total_revenue

# %%
total_orders = df["InvoiceNo"].nunique()
total_orders

# %%
total_customers = df["CustomerID"].nunique()
total_customers

# %%
monthly_sales = (
    df.set_index("InvoiceDate")
      .resample("m")["TotalPrice"]
      .sum()
)

monthly_sales.plot(figsize=(10,5))
plt.title("Aylık toplam satış tutarı")
plt.xlabel("Tarih")
plt.ylabel("Toplam ciro")
plt.show()

# %%
monthly_quantity = (
    df.set_index("InvoiceDate")
      .resample("M")["Quantity"]
      .sum()
)


monthly_quantity.plot(figsize=(10,5))
plt.title("Aylık satılan ürün adedi")
plt.xlabel("Tarih")
plt.ylabel("Toplam adet")
plt.show()

# %%
top_products_qty = (
    df.groupby("Description")["Quantity"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_products_qty.plot(kind="barh", figsize=(8,5))
plt.title("en çok satan 10 ürün (adet)")
plt.xlabel("toplam satılan adet")
plt.ylabel("ürün")
plt.gca().invert_yaxis()
plt.show()

# %%
top_products_rev = (
    df.groupby("Description")["TotalPrice"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)


top_products_rev.plot(kind="barh", figsize=(8,5))
plt.title("en yüksek ciro getiren 10 ürün")
plt.xlabel("toplam ciro")
plt.ylabel("ürün")
plt.gca().invert_yaxis()
plt.show()

# %%
country_sales = (
    df.groupby("Country")["TotalPrice"]
      .sum()
      .sort_values()
      .head(10)
)

country_sales.plot(kind="bar", figsize=(10,5))
plt.title("ülkelere göre toplam satış (ilk 10)")
plt.xlabel("ülke")
plt.ylabel("toplam ciro")
plt.show()

# %%
import pandas as pd
from datetime import timedelta

df = pd.read_csv("online_retail_cleaned.csv")
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

reference_date = df["InvoiceDate"].max() + timedelta(days=1)


rfm = (
    df.groupby("CustomerID")
      .agg({
          "InvoiceDate": lambda  x: (reference_date - x.max()).days,
          "InvoiceNo": "nunique",
          "TotalPrice":"sum"
      })
      .reset_index()
)

rfm.columns = ["CustomerID", "Recency", "Frequency", "Monetary"]
rfm.head()




# %%
rfm["R_Score"] = pd.qcut(rfm["Recency"], 5, labels=[5,4,3,2,1])
rfm["F_Score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1,2,3,4,5])
rfm["M_Score"] = pd.qcut(rfm["Monetary"], 5, labels=[1,2,3,4,5])

rfm["RFM_Score"] = (
    rfm["R_Score"].astype(str) +
    rfm["F_Score"].astype(str) +
    rfm["M_Score"].astype(str)
)

rfm.head()


# %%
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

rfm_scaled = rfm[["Recency", "Frequency", "Monetary"]]

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_scaled)

kmeans = KMeans(n_clusters=4, random_state=42)
rfm["Segment"] = kmeans.fit_predict(rfm_scaled)

rfm.head()


# %%
rfm.groupby("Segment")[["Recency","Frequency","Monetary"]].mean()



```
